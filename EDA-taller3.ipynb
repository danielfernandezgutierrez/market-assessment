```python
#suministro de oficina, tecnologia,mobiliario

#Contando la Cantidad de ordenees por categoria
count_orderCategory= df['Category'].value_counts()
#porcentaje de categoria
percentage = df['Category'].value_counts(normalize=True).head()*100

tabla_freq1 = pd.DataFrame({'Frequency': count_orderCategory,
                               'Percentage %': percentage})
tabla_freq1
```

```python
Furniture_Data=df[df['Category']=='Furniture']
Furniture_Data= Furniture_Data.drop(columns = ['Category'])
Furniture_Data.head(2)
```

```python
Furniture_Data.info()
```

```python
#suministro de oficina, tecnologia,mobiliario

#Contando la Cantidad de ordenees por categoria
count_orderCategory= Furniture_Data['Sub-Category'].value_counts()
#porcentaje de subcategoria categoria
percentage = Furniture_Data['Sub-Category'].value_counts(normalize=True).head()*100
#suma de sub-categoria con el profit
sum_profit = Furniture_Data.groupby(['Sub-Category'])['Profit'].sum().reset_index()
tabla_freq2 = pd.DataFrame({'Frequency': count_orderCategory,
                               'Percentage %': percentage})
tabla_freq2
#Sillas, Mobiliario, Estanterias, Mesas
```
